```python
import pandas as pd

data = pd.read_csv(r"C:\Users\admin\Downloads\test(1).csv")

print("Null values before cleaning:\n", data.isnull().sum())

data['battery_power'].fillna(data['battery_power'].mean(), inplace=True)
data['ram'].fillna(data['ram'].mean(), inplace=True)

data.dropna(inplace=True)

print("\nNull values after cleaning:\n", data.isnull().sum())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\admin\Downloads\test(1).csv")
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['battery_power'], bins=20, kde=True)
plt.title('Distribution of Battery Power')
plt.xlabel('Battery Power (mAh)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['ram'])
plt.title('Box Plot of RAM')
plt.xlabel('RAM (GB)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['ram'], y=data['battery_power'])
plt.title('RAM vs. Battery Power')
plt.xlabel('RAM (GB)')
plt.ylabel('Battery Power (mAh)')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
